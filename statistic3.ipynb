```python
#Q1. What is the Probability density function?

'''[Q1-answer:-]
A probability density function (PDF) is a mathematical function that describes the likelihood of a continuous random variable
falling within a particular range of values. In simpler terms, it gives the probability that a random variable will take on 
a specific value.

For a continuous random variable X, the probability density function f(x) is such that the probability of X being in the
interval [a, b] is given by the integral of f(x) over that interval:

    P(a ≤ X ≤ b) = ∫ab f(x)dx   -----------> # where a and b is denoted the limit of integration

The PDF must satisfy two conditions:

    (1) The function f(x) must be non-negative for all x.
    (2) The total area under the curve of f(x) over the entire range of possible values must be equal to 1.
    
The PDF is a fundamental concept in probability theory and statistics and is used in various applications,
including in the calculation of probabilities, expected values, and other statistical measures.
'''
```

```python
#Q2. What are the types of Probability distribution?

'''[Q2-answer:-
There are several types of probability distributions, each with its own characteristics and applications.
Some common types include:

(1) Discrete Probability Distributions:
        * Bernoulli Distribution: Describes a single trial with two possible outcomes, often denoted as success or failure.
        * Binomial Distribution: Models the number of successes in a fixed number of independent Bernoulli trials.
        * Poisson Distribution: Models the number of events occurring in a fixed interval of time or space, given the average
          rate of occurrence.
                                
(2) Continuous Probability Distributions:
        * Normal (Gaussian) Distribution: One of the most widely used distributions, characterized by a symmetric bell-shaped
          curve.
        * Uniform Distribution: All outcomes within a given interval are equally likely.
        * Exponential Distribution: Describes the time between events in a Poisson process, such as the arrival of customers
          at a service point.
        * Gamma Distribution: Generalizes the exponential distribution and is used to model waiting times or time until a 
          given number of events occur.
        * Beta Distribution: Often used to model probabilities or proportions.
        
(3) Multinomial Distribution: Generalizes the binomial distribution to more than two categories.

(4) Hypergeometric Distribution: Models the number of successes in a sequence of draws from a finite population without
    replacement.

(5) Geometric Distribution: Models the number of trials needed until the first success in a sequence of independent
    Bernoulli trials

'''
```

```python
#Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

'''[Q3-answer:-]
To calculate the probability density function (PDF) of a normal distribution at a given point x with a
specified mean μ and standard deviation σ, you can use the formula for the normal distribution PDF:

                f(x)= 1 / (σ*(2π)^0.5) exp(−((x−μ)^2)/2σ^2) 
'''

import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
        x (float): The point at which to calculate the PDF.
        mean (float): The mean of the normal distribution.
        std_dev (float): The standard deviation of the normal distribution.

    Returns:
        float: The PDF value at the given point.
    """
    exponent = -((x - mean) ** 2) / (2 * (std_dev ** 2))
    pdf = (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(exponent)
    return pdf

# Example usage
mean = 0
std_dev = 1
x = 1
pdf_value = normal_pdf(x, mean, std_dev)
print("PDF at x =", x, ":", pdf_value)
```

```python
#Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

'''[Q4-answer:-]
The properties of the Binomial distribution are as follows:

    (1) Fixed number of trials (n): The number of trials or experiments is fixed in advance.
    (2) Independent trials: Each trial or experiment is independent of the others.
    (3) Two outcomes: Each trial results in one of two outcomes, often termed "success" and "failure".
    (4) Constant probability of success (p): The probability of success (denoted by p) remains constant from trial to trial.
    (5) Discrete distribution: The distribution is discrete, meaning that it deals with a finite number of outcomes.

    The Binomial distribution is typically denoted as B(n, p), where 'n' represents the number of trials and 'p' represents
    the probability of success on each trial.

----> Two examples of events where the Binomial distribution can be applied are:

            (1) Coin Flipping: Consider flipping a fair coin (where the probability of getting heads is 0.5) 10 times.
                Each flip is independent of the others, and there are only two possible outcomes (heads or tails). 
                The number of heads obtained in these 10 flips follows a Binomial distribution with parameters n = 10 
                (number of trials) and p = 0.5 (probability of success - getting heads).

            (2) Exam Success: Suppose a student takes a 10-question multiple-choice exam where each question has four choices,
                and they answer randomly without knowledge of the material. Let's say the probability of answering a
                question correctly by random guessing is 0.25. The number of correct answers the student gets follows
                a Binomial distribution with parameters n = 10 (number of questions) and p = 0.25 
                (probability of success - answering a question correctly by random guessing).
'''
```

```python
#Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

'''[Q6-answer:-]'''

import math

def poisson_cdf(mean, k):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution.
    
    Parameters:
        mean (float): The mean of the Poisson distribution.
        k (int): The point at which to calculate the CDF.
        
    Returns:
        float: The CDF value at the given point.
    """
    if k < 0:
        return 0.0  # Cumulative probability cannot be negative
    else:
        cdf = 0.0
        for i in range(k + 1):
            cdf += (math.exp(-mean) * (mean ** i)) / math.factorial(i)
        return cdf

# Example usage
mean = 5
k = 3
cdf_value = poisson_cdf(mean, k)
print("CDF at k =", k, ":", cdf_value)
```

```python
#Q7. How Binomial distribution different from Poisson distribution?

'''[Q7-answer:-]
Binomial and Poisson distributions are both probability distributions used to model the number of successes or events in 
a fixed number of trials or a fixed interval of time. However, they differ in their assumptions and the scenarios they
are typically used to model:

    (1) Assumptions:
            (a) Binomial Distribution: Assumes a fixed number of trials (n), with each trial having two possible outcomes 
                (success or failure) and a constant probability of success (p) for each trial. The trials are independent of 
                each other.
            (b) Poisson Distribution: Assumes an interval of time or space, and counts the number of events that occur within
                that interval. It's often used when the number of trials is large, but the probability of success per trial
                is very small. The events are assumed to occur independently, but the probability of an event occurring in a
                very small sub-interval is proportional to the length of the sub-interval.
                
    (2) Nature of Events:
            (a) Binomial Distribution: Typically used to model discrete events where there are a fixed number of trials,
                each with two possible outcomes (success or failure), such as coin flips, pass/fail outcomes, etc.
            (b) Poisson Distribution: Used to model the number of events occurring within a fixed interval of time or space, 
                such as the number of phone calls received in an hour, the number of arrivals at a bus stop in a given time
                period, etc.
                
    (3) Parameters:
            (a) Binomial Distribution: Governed by two parameters: n (the number of trials) and p (the probability of success).
            (b) Poisson Distribution: Governed by one parameter, λ (lambda), which represents the average rate of event
                occurrences within the fixed interval.
                
    (4) Probability Mass Function (PMF):
            (a) Binomial Distribution: The PMF calculates the probability of getting a specific number of successes in n 
                trials.
            (b) Poisson Distribution: The PMF calculates the probability of observing a specific number of events within a
                fixed interval of time or space.
'''
```

```python
#Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

'''[Q8-answer:-]'''

import numpy as np
#parameter
mean = 5
sample_size = 1000

#generate random sample of poisson distribution
sample = np.random.poisson(mean,size=sample_size)

#calculate the sample mean and variance
mean = np.mean(sample)
variance = np.var(sample)

#print result
print('sample mean',mean)
print('sample variance',variance)
```

```python
#Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

'''[Q9-answer:-]
In both the Binomial and Poisson distributions, the mean (expected value) and variance are closely related, but their exact
relationship differs slightly between the two distributions:

    (1) Binomial Distribution:
            * Mean: The mean (expected value) of a Binomial distribution with parameters n (number of trials) and
                    p (probability of success) is given by μ = np.
            * Variance: The variance of a Binomial distribution is given by σ^2 = np(1 - p).
            
    In the Binomial distribution, the variance is directly dependent on both the number of trials (n) and the probability of
    success (p). As the number of trials increases, the variance tends to increase proportionally, assuming the probability of
    success remains constant. Additionally, the variance is also influenced by the spread of probabilities around the mean.

    (2) Poisson Distribution:
            * Mean: The mean (expected value) of a Poisson distribution with parameter λ (lambda) is given by μ = λ.
            * Variance: The variance of a Poisson distribution is also given by σ^2 = λ.
            
    In the Poisson distribution, the variance is equal to the mean. This is a unique property of the Poisson distribution.
    The variance of a Poisson distribution is independent of any other parameters, as it is solely determined by the mean λ.
    When λ increases, the spread of the distribution increases as well, but the ratio of variance to mean remains constant.
'''
```
